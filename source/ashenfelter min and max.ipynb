```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from scipy.optimize import minimize, Bounds
import pandas as pd
from tabulate import tabulate

df2 = pd.read_excel(r"C:\Users\polin\Dropbox\EE 608 spring 2023\project\Wine-Quality-Modeling\data\bordeauxWeather.xlsx")
Year2 = df2['year'].values.tolist()
avgTempSept = df2['sepT'].values.tolist()
avgTempGrowing = df2['avgTempAprilSeptember'].values.tolist()
totalPrecipAug = df2['totalPSeptAug'].values.tolist()
totalPrecipWinter = df2['totalPOctMar'].values.tolist()

year=np.array(Year2)
TempAprSep=np.array(avgTempGrowing)
AugPrecip=np.array(totalPrecipAug)
WinterPrecip=np.array(totalPrecipWinter)
SepTempAvg=np.array(avgTempSept)

minage=0
maxage=100

maxTempSep=max(SepTempAvg)
minTempSep=min(SepTempAvg)

maxTempAprSep=max(TempAprSep)
minTempAprSep=min(TempAprSep)

maxPrecipAug=max(totalPrecipAug)
minPrecipAug=min(totalPrecipAug)

maxPrecipWinter=max(totalPrecipWinter)
minPrecipWinter=min(totalPrecipWinter)

def f(x):
    fx=-(.608*(x[0]) - .00380*(x[1]) + .00115*(x[2]) + .00765*(x[3]))
    return fx

def g(x):
    gx=-(.240*(x[0])+.608*(x[1]) - .00380*(x[2]) + .00115*(x[3]) + .00765*(x[4]))
    return gx


x_start=np.array([1, 1, 1, 1])

x_start2=np.array([1, 1, 1, 1, 1])


lim = [minTempAprSep, minPrecipAug, minPrecipWinter, minTempSep, maxTempAprSep, maxPrecipAug, maxPrecipWinter, maxTempSep]
lim2 = [minage, minTempAprSep, minPrecipAug, minPrecipWinter, minTempSep, maxage, maxTempAprSep, maxPrecipAug, maxPrecipWinter, maxTempSep]

bounds = Bounds([lim[0], lim[1], lim[2], lim[3]],  # [min x0, min x1]
                [lim[4], lim[5], lim[6], lim[7]])  # [max x0, max x1]

bounds2 = Bounds([lim2[0], lim2[1], lim2[2], lim2[3], lim2[4]],  # [min x0, min x1]
                [lim2[5], lim2[6], lim2[7], lim2[8], lim2[9]])  # [max x0, max x1]

result = minimize(f, x_start, bounds=bounds)
result2 = minimize(g, x_start2, bounds=bounds2)

result.x.round(3)
result2.x.round(3)

print("Max Quality,w/o age:",abs(result.fun))
print("Max Quality w/ age:",abs(result2.fun))
print("Points of Max w/o age:",result.x)
print("Points of Max w/ age:",result2.x)
print("MaxTempAprSep:",maxTempAprSep)
print("MinTempAprSep:",minTempAprSep)
print("MaxPrecipAug:",maxPrecipAug)
print("MinPrecipAug:",minPrecipAug)
print("Maxprecipwint:",maxPrecipWinter)
print("Minprecipwint:",minPrecipWinter)
print("MaxTempSep:",maxTempSep)
print("MinTempSep:",minTempSep)
```
